```python
# to read and visualize spatial data
import geopandas as gpd
import matplotlib.pyplot as plt 
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
path = "./tl_2021_42_tract/tl_2021_42_tract.shp"
df = gpd.read_file(path)
df = df.to_crs("EPSG:4326")
```

```python
df.head()
```

```python
df.shape
```

```python
df_2 = gpd.read_file("FoodAccessResearchAtlasData2019_.csv") 
```

```python
df_2.head()
```

```python
df_2.shape
```

```python
import pandas as pd
# merge, key=CensusTract
df_3 = pd.merge(df, df_2, left_on='GEOID', right_on='CensusTract')
```

```python
df_3.head()
```

```python
df_3.shape
```

```python
df_3[["GEOID", "CensusTract", "LILATracts_1And10"]]
```

```python
df[["geometry"]]
```

```python
df_4 = pd.merge(df_3[["GEOID", "CensusTract", "LILATracts_1And10"]], 
                df[["GEOID", "ALAND", "AWATER", "INTPTLAT", "INTPTLON", "geometry"]], on="GEOID")
```

```python
df_4
```

```python
df_4.plot()
```

```python
df.plot()
```

```python
gdf_4 = gpd.GeoDataFrame(df_4, geometry='geometry') 
```

```python
gdf_4.plot()
```

```python
f,ax = plt.subplots(1,1, figsize=(8,6), sharex=True, sharey=True, dpi=300)
plt.title('Simple Map')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="3%",pad=0,alpha=0.5)
gdf_4.plot('LILATracts_1And10', ax=ax, alpha=0.5, cmap='Pastel1', edgecolor='k', legend=True, cax=cax, linewidth=0.1)
plt.show()
```

```python
df_5 = gpd.read_file("FoodAccessResearchAtlas_.csv") 
```

```python
import pandas as pd
# merge, key=CensusTract
df_6 = pd.merge(df, df_5, left_on='GEOID', right_on='CensusTract')
```

```python
df_6 = pd.merge(df_6[["GEOID", "CensusTract", "LILATracts_1And10"]], 
                df[["GEOID", "geometry"]], on="GEOID")
```

```python
gdf_6 = gpd.GeoDataFrame(df_6, geometry='geometry') 
```

```python
f,ax = plt.subplots(1,1, figsize=(8,6), sharex=True, sharey=True, dpi=300)
plt.title('LILATracts_1And10 in the 11 counties')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="3%",pad=0,alpha=0.5)
gdf_6.plot('LILATracts_1And10', ax=ax, alpha=0.5, cmap='Pastel1', edgecolor='k', legend=True, cax=cax, linewidth=0.1)
plt.show()
```

```python
f,ax = plt.subplots(1,1, figsize=(8,6), sharex=True, sharey=True, dpi=300)
plt.title('LILATracts_1And10 in the 11 counties (showing food insecurity area)')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="3%",pad=0,alpha=0.5)
gdf_6.plot('LILATracts_1And10', ax=ax, alpha=0.5, cmap='bwr', edgecolor='k', legend=True, cax=cax, linewidth=0.1)
plt.show()
```
